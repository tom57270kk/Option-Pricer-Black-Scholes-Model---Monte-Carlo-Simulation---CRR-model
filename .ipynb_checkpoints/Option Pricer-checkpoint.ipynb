# 1. Geometric Brownian Motion

Black-Scholes model assumption :  $X_t$ follows the GBM : 

$dX_{t} = \mu X_{t} dt + \sigma X_{t} dB_{t}$

We implement the GBM knowing its explicit solution : $X_{t} = X_{0} e^{\mu t + \sigma B_{t} - \frac{\sigma^2}{2} t}$

```python
import numpy as np
import matplotlib.pyplot as plt

def geometric_brownian_motion(mu, sigma, S0, T, N, seed=None):
    
    if seed is not None:
        np.random.seed(seed)
        
    dt = T / N
    t = np.linspace(0., T, N+1)
    W = np.random.standard_normal(size=N+1)
    W = np.cumsum(W) * np.sqrt(dt)
    X = (mu - 0.5 * sigma**2) * t + sigma * W
    S = S0 * np.exp(X)
    return S

# paramtres
S0 = 20  # valeur initiale
mu = 0.05   # rendement moyen
sigma = 0.06 # volatilit
T = 1.0     # dure de la simulation en annes
N = 252     # nombre de pas de temps (jours dans une anne de trading)

# Simulation
gbm = geometric_brownian_motion(mu, sigma, S0, T, N)
print(gbm[-1])
# Plot
plt.plot(gbm)
plt.xlabel('Time')
plt.ylabel('PRice')
plt.title('GBM Simulation')
plt.grid(True)
plt.show()
```

## 2.2. Explicit Black-Scholes Formula

```python
#SOLUTION EXPLICITE DU GBM POUR UNE CALL OPTION

def bs_call_option(x0,K,sigma,T,mu) :

    d1 = (np.log(x0/K)+mu)/(sigma*np.sqrt(T)) + sigma*np.sqrt(T)*0.5
    
    d2 = d1 - sigma*np.sqrt(T)
    
    return x0*norm.cdf(d1) - K*np.exp(-mu*T)*norm.cdf(d2)



#AFFICHE LE PAYOFF EN FONCTION DU PRIX DE L'ACTIF SOUS JACENT AVEC LA FORMULE EXPLICITE

def payoffX(K,sigma,mu,T):
    X = np.linspace(1,100,1000)
    Y = [bs_call_option(x,K,sigma,T,mu) for x in X]
    #plt.figure()
    plt.plot(X,Y,label="Black-Scholes Formula")
    plt.xlabel("Stock Price")
    plt.ylabel("Payoff")
    plt.legend()
    

#AFFICHE LE PAYOFF EN FONCTION DE LA MATURITE AVEC LA FORMULE EXPLICITE

def payoffT(X0,K,sigma,mu,T):
    Te = [k for k in range(1,T)]
    Ye = [bs_call_option(X0,K,sigma,k,mu) for k in range(1,T)]
    plt.figure()
    plt.plot(Te,Ye,label="Explicite")
    plt.xlabel("Time t")
    plt.ylabel("Payoff")
    plt.legend()
    
    #return Te,Ye




```

## 2.4 Variance Reduction

We can remark :

\begin{equation*}
    (X_t - K)^+= (K - X_t)^+ + X_t - K
\end{equation*}

We can also compute that :

$\mathbb{E}[X_T - K] = X_0 e^{\mu T} - K$

Then, we implement it

```python
def monte_carlo(S0,mu,sigma,K,T,N):
    S = np.zeros(N)
    for k in range (N):
        S[k] = geometric_brownian_motion(mu, sigma, S0, T, N, seed=None)[-1]

    payoff1 = np.maximum(K -S , 0)
    payoff1 = np.mean(payoff1)
    payoff2 = S0 * np.exp(mu*T ) - K
    payoff = payoff1 + payoff2
    return np.exp(-mu*T) * payoff
```

## 2.6 Implied Volatility

By looking at implied volatility, we can observe Black Scholes model doesn't hold in reality.

```python

import numpy as np
import matplotlib.pyplot as plt

def plot_implied_volatility(symbol, expiry_date):
    # on charge les donnes
    option_chain = yf.Ticker(symbol).option_chain(expiry_date)
    calls = option_chain.calls
    
    # pour les calls, on prend le stike et la vol imp
    strikes = calls['strike']
    implied_vols = calls['impliedVolatility']
    
    
    plt.plot(strikes, implied_vols, marker='o', linestyle='-')
    plt.xlabel('Strike Price')
    plt.ylabel('Implied Volatility')
    plt.title(f'Implied Volatility for {symbol} Options ({expiry_date})')
    plt.grid(True)
    plt.show()


symbol = 'GOOG'
expiry_date = '2024-06-14'   
plot_implied_volatility(symbol, expiry_date)
```

# 3. CRR model

Black-Scholes model provides us a framework to model price evolution of derivatives whose value is based on the value of an underlying asset $x$ and which has a maturity $T$. However, that type of derivatives is very restrictive. \\
That's why there exists other models more adapted to derivatives which do not have a fixed  maturity $T$ such as American options\footnote{Similar to European options but with the right to sell/buy at any time}.

\begin{equation*}
    p = \frac{e^{(r \, \Delta_t)} - d}{u-d}
\end{equation*}

\begin{equation*}
    u = \exp(\sigma \sqrt{\Delta_t})
\end{equation*}

\begin{equation*}
    d = \frac{1}{u}
\end{equation*}

## 3.1 CRR model : European option

```python
import numpy as np

def crr_call_option(S, K, T, r, sigma, n):
    dt = T / n
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)
    
    # calcul des prix des actions  chaque tape
    stock_price = np.zeros((n+1, n+1))
    for j in range(n+1):
        for i in range(j+1):
            stock_price[i, j] = S * (u ** (j-i)) * (d ** i)
    
    # calcul des valeurs d'option  l'chance
    option_value = np.maximum(stock_price[:, n] - K, 0)
    
    # rtro-propagation pour calculer le prix de l'option
    for j in range(n-1, -1, -1):
        for i in range(j+1):
            option_value[i] = np.exp(-r * dt) * (p * option_value[i] + (1 - p) * option_value[i+1])
    
    return option_value[0]


S = 100      
K = 50      
T = 1       
r = 0.01    
sigma = 0.2 
n = 10    
# Calcul du prix de l'option d'achat
call_price = crr_call_option(S, K, T, r, sigma, n)
print("Le prix de l'option d'achat europenne est:", call_price)
```

## 3.3 CRR model for european option : Graphic version

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

def crr_call_option(S, K, T, r, sigma, n):
    dt = T / n
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)
    
    # Stock price tree with simulated prices
    stock_price_tree = nx.DiGraph()
    for j in range(n+1):
        for i in range(j+1):
            price = S * (u ** (j-i)) * (d ** i)
            stock_price_tree.add_node((i, j), pos=(i, -j), price=price)
            if j < n:
                stock_price_tree.add_edge((i, j), (i, j+1))
                stock_price_tree.add_edge((i, j), (i+1, j+1))
    
    # Option value tree
    option_value_tree = nx.DiGraph()
    for j in range(n+1):
        for i in range(j+1):
            option_value_tree.add_node((i, j), pos=(i, -j))
            if j == n:
                option_value_tree.nodes[(i, j)]['value'] = max(stock_price_tree.nodes[(i, j)]['price'] - K, 0)
            if j < n:
                option_value_tree.add_edge((i, j), (i, j+1))
                option_value_tree.add_edge((i, j), (i+1, j+1))
    
    # Calculate option values using backward induction
    for j in range(n-1, -1, -1):
        for i in range(j+1):
            option_value_tree.nodes[(i, j)]['value'] = np.exp(-r * dt) * (
                p * option_value_tree.nodes[(i, j+1)]['value'] + 
                (1 - p) * option_value_tree.nodes[(i+1, j+1)]['value'])
    
    return stock_price_tree, option_value_tree

# Parameters
S = 100      # Current price of the underlying asset
K = 110      # Strike price
T = 1        # Time to maturity in years
r = 0.01    # Risk-free interest rate
sigma = 0.2  # Volatility
n = 5        # Number of steps in the binomial tree

# Calculate option price and create trees
stock_price_tree, option_value_tree = crr_call_option(S, K, T, r, sigma, n)

# Plot stock price tree with simulated prices
plt.figure(figsize=(10, 5))
pos = nx.get_node_attributes(stock_price_tree, 'pos')
prices = nx.get_node_attributes(stock_price_tree, 'price')
nx.draw(stock_price_tree, pos, with_labels=False, node_size=500, node_color='lightblue', font_size=10)
nx.draw_networkx_labels(stock_price_tree, pos, labels=prices, font_size=10)
plt.title('Stock Price Tree')
plt.show()

# Plot option value tree
plt.figure(figsize=(10, 5))
pos = nx.get_node_attributes(option_value_tree, 'pos')
nx.draw(option_value_tree, pos, with_labels=False, node_size=500, node_color='lightgreen', font_size=10)
labels = nx.get_node_attributes(option_value_tree, 'value')
nx.draw_networkx_labels(option_value_tree, pos, labels=labels, font_size=10)
plt.title('Call Option Value Tree')
plt.show()
```

## 3.4 CRR model : American Option

Same as european call option except that : 

\begin{equation}
    V(i, j) = \max \left( \exp(-r \cdot dt) \cdot \left[ p \cdot V(i, j+1) + (1-p) \cdot V(i+1, j+1) \right], \text{Payoff}(i, j) \right)
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

def crr_american_option(S, K, T, r, sigma, n):
    dt = T / n
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)
    
    # Stock price tree with simulated prices
    stock_price_tree = nx.DiGraph()
    for j in range(n+1):
        for i in range(j+1):
            price = S * (u ** (j-i)) * (d ** i)
            stock_price_tree.add_node((i, j), pos=(i, -j), price=price)
            if j < n:
                stock_price_tree.add_edge((i, j), (i, j+1))
                stock_price_tree.add_edge((i, j), (i+1, j+1))
    
    # Option value tree
    option_value_tree = nx.DiGraph()
    for j in range(n+1):
        for i in range(j+1):
            option_value_tree.add_node((i, j), pos=(i, -j))
            if j == n:
                option_value_tree.nodes[(i, j)]['value'] = max(stock_price_tree.nodes[(i, j)]['price'] - K, 0)
            if j < n:
                option_value_tree.add_edge((i, j), (i, j+1))
                option_value_tree.add_edge((i, j), (i+1, j+1))
    
    # Calculate option values using backward induction
    for j in range(n-1, -1, -1):
        for i in range(j+1):
            immediate_exercise_value = max(stock_price_tree.nodes[(i, j)]['price'] - K, 0)
            option_value_tree.nodes[(i, j)]['value'] = np.exp(-r * dt) * (
                p * option_value_tree.nodes[(i, j+1)]['value'] + 
                (1 - p) * option_value_tree.nodes[(i+1, j+1)]['value'])
            option_value_tree.nodes[(i, j)]['value'] = max(option_value_tree.nodes[(i, j)]['value'], immediate_exercise_value)
    
    return stock_price_tree, option_value_tree

# Parameters
S = 100      # Current price of the underlying asset
K = 50      # Strike price
T = 1       # Time to maturity in years
r = 0.01     # Risk-free interest rate
sigma = 0.2  # Volatility
n = 5        # Number of steps in the binomial tree

# Calculate option price and create trees
stock_price_tree, option_value_tree = crr_american_option(S, K, T, r, sigma, n)

# Plot stock price tree with simulated prices
plt.figure(figsize=(10, 5))
pos = nx.get_node_attributes(stock_price_tree, 'pos')
prices = nx.get_node_attributes(stock_price_tree, 'price')
nx.draw(stock_price_tree, pos, with_labels=False, node_size=500, node_color='lightblue', font_size=10)
nx.draw_networkx_labels(stock_price_tree, pos, labels=prices, font_size=10)
plt.title('Stock Price Tree with Simulated Prices')
plt.show()

# Plot option value tree
plt.figure(figsize=(10, 5))
pos = nx.get_node_attributes(option_value_tree, 'pos')
nx.draw(option_value_tree, pos, with_labels=False, node_size=500, node_color='lightgreen', font_size=10)
labels = nx.get_node_attributes(option_value_tree, 'value')
nx.draw_networkx_labels(option_value_tree, pos, labels=labels, font_size=10)
plt.title('American Call Option Value Tree')
plt.show()
```
